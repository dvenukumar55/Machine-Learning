```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
data = {
    'hours_studied': [1, 2, 3, 4, 5, 6, 7, 8],
    'sleep_hours': [8, 7, 6, 5, 5, 4, 3, 2],
    'passed': [0, 0, 0, 0, 1, 1, 1, 1]
}

df = pd.DataFrame(data)
df
```

```python
X = df[['hours_studied', 'sleep_hours']]
y = df['passed']
model = GaussianNB()
model.fit(X, y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create mesh grid
x_min, x_max = X['hours_studied'].min()-1, X['hours_studied'].max()+1
y_min, y_max = X['sleep_hours'].min()-1, X['sleep_hours'].max()+1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1),
                     np.arange(y_min, y_max, 0.1))

Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
plt.scatter(X['hours_studied'], X['sleep_hours'], c=y, cmap='coolwarm', edgecolors='k')
plt.xlabel('Hours Studied')
plt.ylabel('Sleep Hours')
plt.title('Naive Bayes Classification Boundary')
plt.show()
```

```python
new_student = [[6,3]]
prediction = model.predict(new_student)
probability = model.predict_proba(new_student)

print("Predicted (0=Fail, 1=Pass):", prediction[0])
print("Probabilities [Fail, Pass]:", probability[0])
```
